### If-Statements with user defined functions

```python
# Defining price label function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Applying price label function on dataframe
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

# Task 4.7

```python
# Step 2 Creating busiest-days column with for-loop function
result = []

for value in df_ords_prods['orders_day_of_week']:
  if value == 0 or value == 1:
    result.append('Busiest days')
  elif value == 3 or value == 4:
    result.append('Slowest days')
  else:
    result.append('Regularly busy')
```

```python
# Print result
result
```

```python
# Creating new column for results of for-loop function
df_ords_prods['busiest_days'] = result
```

```python
# Check distribution
df_ords_prods['busiest_days'].value_counts(dropna = False)
```

#### As expected the number of 'busiest days' becomes even closer to the number of 'regularly busy' days and 'slowest days' also increases. All the values add up to same number as the last column so there are no inconsistencies there.

```python
# Check busiest period of day
df_ords_prods['order_hour_of_day'].value_counts(dropna = False)
```

```python
# Step 4 Creating busiest period of day with for-loop function
result = []
for x in df_ords_prods['order_hour_of_day']:
    if x in {10, 11, 13, 14, 15}:
        result.append('Most orders')
    elif x in {1, 2, 3, 4, 5}:
        result.append('Fewest orders')
    else:
        result.append('Average orders')
```

```python
# Print result
result
```

```python
# Creating new column for results of for-loop function
df_ords_prods['busiest_period_of_day'] = result
```

```python
# Checking distribution of new column
df_ords_prods['busiest_period_of_day'].value_counts(dropna = False)
```
