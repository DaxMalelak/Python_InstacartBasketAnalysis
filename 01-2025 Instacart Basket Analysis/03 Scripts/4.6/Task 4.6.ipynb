#### 2. Merging Dataframes

```python
# Dropping _merge column from df_ords_prods_comb
df_ords_prods_comb = df_ords_prods_comb.drop(columns=['_merge'])
```

```python
# Merging orders_products_combined.pkl and products_checked.csv
df_merged = df_ords_prods_comb.merge(df_prods, on = 'product_id', indicator = True)
```

```python
# Check output of df_merged
df_merged.head()
```

```python
# Removing unnamed columns
df_merged = df_merged.loc[:, ~df_merged.columns.str.contains('^Unnamed')]
```

```python
# Check output of df_merged
df_merged.head()
```

```python
# Check shape of df_merged
df_merged.shape
```

```python
# Step 6 Check merge rates
df_merged['_merge'].value_counts()
```

#### 3. Exporting Newly Created Dataframe

```python
# Exporting df_merged as pkl file
df_merged.to_pickle(os.path.join(path, '02 Data','Prepared Data', 'ords_prods_merged.pkl'))
```
